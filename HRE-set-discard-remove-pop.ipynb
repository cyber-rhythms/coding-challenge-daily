```python
n = int(input())
s = set(map(int, input().split()))
N = int(input())

for _ in range(N):
    cmds = input().split()

    # Process the command in the list 
    method = cmds[0]
    if len(cmds) == 2:
        arg = cmds[1]
        cmd = f"s.{method}({arg})"
    else:
        cmd = f"s.{method}()"

    # Calling `remove("x")` may result in Key Error. 
    try:
        _ = exec(cmd)
    except KeyError:
        continue 

print(sum(s))
```

Code Review.

Deficiencies.

* Array indexing is not robust to out of range index errors - slicing is -- removing the need to code your if and else blocks. Slicing out of range yields an empty list, which is useful for `getattr(obj, method)(*args)`.
* Intead of `exec()`, use `getattr(obj, method)`, which returns a callable object, which we can then call using `getattr(obj, method)(*args)`.
* `exec()` makes Python parse and compile a string into bytecode at runtime - slow.

Solution.

* Rewritten using slicing to remove case handling.
* Note that you are unpacking the iterator constructed by `map()`

```python
n = int(input())
s = set(map(int, input().split()))
N = int(input())

for _ in range(N):
    line = input().split()

    command = line[0]
    args = map(int, line[1:])

    try:
        getattr(s, command)(*args)
    except KeyError:
        continue

print(sum(s))
```
