Write Number in Expanded Form
You will be given a number and you will need to return it as a string in Expanded Form. For example:

   12 --> "10 + 2"
   45 --> "40 + 5"
70304 --> "70000 + 300 + 4"
NOTE: All numbers will be whole numbers greater than 0.

Ideas.

* Failure modes/constraints: We always get a whole number greater than 0.
* Standard practice for digit representation is that $abcd = a \times 10^3 + \dots d \times 10^0.$
* Just code this rule and then string format - don't worry so much about conciseness prematurely, just get it out naturally.

```python
# Final submission.

def expanded_form(num):
    num = str(num)
    expanded = []
    for i in range(len(num)):
        base = 10 ** (len(num) - 1 - i)
        num_base = str(int(num[i]) * base)
        expanded.append(num_base)

    expanded = [element for element in expanded if element != "0"]

    return " + ".join(expanded)

expanded_form(1001)
```
