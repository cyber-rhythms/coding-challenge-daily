Ideas.

* If we let $\tilde{M}$ be a scalar random variable with uniform probability mass function $P(\tilde{M} = M_{i,j}) = 1 / (r+1)(c+1)$, then with $M_{i, j} = (i+j)^2$,

$$\mathbb{E}[\tilde{M}] = \sum^r_{i=0} \sum^{c}_{j=0} \frac{(i+j)^2}{(r+1)(c+1)}.$$
* Noting that the double summation is on the order $O(rc)$, and given that $0 \leq r, c, \leq 10^{18}$, and that even a single summation will be either $O(r)$ or $O(c)$, suggests the problem is infeasible without an analytic method.
* The sum of the first $n$ integers has formula,

$$\sum^n_{i=1} i = \frac{n(n+1)}{2}.$$
* The sum of squares of the first $n$ integers has formula,

$$\sum^n_{i=1} i^2 = \frac{n(n+1)(2n+1)}{6}.$$
* Via distributive and associative properties of summation,

$$\sum_i \sum_j M_{ij} = \sum_i \sum_j i^2 + 2ij + j^2 = (c+1)\left( \sum_i i^2\right) + 2\left(\sum_i i\right)\left(\sum_j j\right) + (r+1)\left(\sum_j j^2\right).$$
* Use `Fractions` module and force an integer divide using `//` to avoid loss of precision on floats, which lose precision after 15-17 decimal digits. 

```python
# Submission.

from fractions import Fraction

def integer_sum(n):
    return n * (n + 1) // 2

def square_sum(n):
    return (n * (n + 1) * (2*n + 1)) // 6

def expected_value(r, c):
    denom = ((r + 1)*(c + 1))
    num = (((c + 1) * square_sum(r)) 
            + (2 * integer_sum(r) * integer_sum(c))
            + ((r + 1) * square_sum(c)))
    return Fraction(num, denom)
```
