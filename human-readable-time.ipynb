Write a function, which takes a non-negative integer (seconds) as input and returns the time in a human-readable format (HH:MM:SS)

HH = hours, padded to 2 digits, range: 00 - 99
MM = minutes, padded to 2 digits, range: 00 - 59
SS = seconds, padded to 2 digits, range: 00 - 59
The maximum time never exceeds 359999 (99:59:59)

You can find some examples in the test fixtures.

Ideas.

* I have a vague recollection that time is base 60, but I can't remember how to convert between bases. Work it out, no need for lookup.
* Let's say we have some number in seconds, $x = 7198$.
* A human based formula is given below, but now need to make that logic explicit by dealing with other cases.
* Do we program minutes and seconds recursively, or absolutely -> which will be easier for running the logic?


```python
# HH -> 7198 // (60 ** 2) -> x // (60 ** 2)
print(7198 / (60 ** 2))
print(7198 % (60 ** 2))

## Minutes -> (x % 60 ** 2) // (60 ** 1)
print(3598 / ( 60 ** 1))
print(3598 % (60 ** 1))

## Seconds -> (x % 60 ** 2) % (60 ** 1)
print(3598 / ( 60 ** 1))
print(3598 % (60 ** 1))
```

```python
# Approach 1: Too verbose, no need for mathematical powers, even though it reminds you of a base problem.

x = 3599
h, h_r = x // (60 ** 2), x % (60 ** 2)
m, m_r = h_r // (60 ** 1), h_r % (60 ** 1)
s = m_r

time = [h, m, s]
fmt_time = []
for num in time:
    if len(str(num)) != 2:
        fmt_time.append("".join(["0", str(num)]))
    else:
        fmt_time.append(str(num))
fmt_time
```

```python
# Approach 2: Too condensed and difficult to read.

time = [x // (60 ** 2), (x % (60 ** 2)) // (60 ** 1), (x % 60 ** 2) % (60 ** 1) ]
":".join(["".join(["0", str(num)]) if len(str(num)) != 2 else str(num) for num in time])
```

```python
# Approach 3: Favoured as more readable, and also, you wrap the if/else and ugly string joining logic into f-strings, which is more readable.
# Note that remainder divide by 3600 then remainder divide by 60 is same as remainder divide by 60.

def make_readable(seconds):
    h = seconds // 3600
    m = (seconds % 3600) // 60
    s = seconds % 60
    return f"{h:02}:{m:02}:{s:02}"
```
