Given two arrays a and b write a function comp(a, b) (orcompSame(a, b)) that checks whether the two arrays have the "same" elements, with the same multiplicities (the multiplicity of a member is the number of times it appears). "Same" means, here, that the elements in b are the elements in a squared, regardless of the order.

Examples
Valid arrays
a = [121, 144, 19, 161, 19, 144, 19, 11]  
b = [121, 14641, 20736, 361, 25921, 361, 20736, 361]
comp(a, b) returns true because in b 121 is the square of 11, 14641 is the square of 121, 20736 the square of 144, 361 the square of 19, 25921 the square of 161, and so on. It gets obvious if we write b's elements in terms of squares:

a = [121, 144, 19, 161, 19, 144, 19, 11] 
b = [11*11, 121*121, 144*144, 19*19, 161*161, 19*19, 144*144, 19*19]
Invalid arrays
If, for example, we change the first number to something else, comp is not returning true anymore:

a = [121, 144, 19, 161, 19, 144, 19, 11]  
b = [132, 14641, 20736, 361, 25921, 361, 20736, 361]
comp(a,b) returns false because in b 132 is not the square of any number of a.

a = [121, 144, 19, 161, 19, 144, 19, 11]  
b = [121, 14641, 20736, 36100, 25921, 361, 20736, 361]
comp(a,b) returns false because in b 36100 is not the square of any number of a.

Ideas.

* Data structures - list or dictionary? Not quite sure about the interaction between multiplicities and whether all numbers in $B$ are squares of $A$.
* Compute the squares of list $A$, and also track their multiplicities, so use a "square dictionary.
* Check whether every number in list $B$ is both in $A$ and has the same multiplicity.
* Failure mode 1: Any array that is empty will need an exit False.
* Failure mode 2: If list $B$ contains any number that isn't square, then number $B$ won't be in the "square dicitonary".
* Possible failure mode 3: Could there be an example where every number in list $B$ is a square of some number in $A$, but we fail the multiplicities test? If arrays are of same length then this failure mode doens't appear possible, but will need a fail-safe for this as question doesn't specify- use dicitonaries.
* This time I won't use collections.Counter, but acknowledge in Code review, as even though latter is better for SWE, not sure if it's allowed any more in CodeWars?

```python
# Final submission.

def comp(array1, array2):
    
    # Only return False if one of the two arrays is empty. 
    if array1 is None or array2 is None:
        return False

    # Dictionary of squares of list A and counts.
    squares_count = dict()
    for num in array1:
        square = num ** 2
        squares_count[square] = squares_count.get(square, 0) + 1

    # Dictionary of list B and counts.
    array2_count = dict()
    for num in array2:
        if not squares_count.get(num, 0):
            return False
        else:
            array2_count[num] = array2_count.get(num, 0) + 1
    
    # Compare multiplicities (as not sure whether arrays are of same length).
    return squares_count == array2_count
```

```python
# Final submission.

from collections import Counter

def comp(array1, array2):
    
    # Only return False if one of the two arrays is empty. 
    if array1 is None or array2 is None:
        return False

    squares_count = Counter(num ** 2 for num in array1)
    array2_count = Counter(array2)
    
    # Compare multiplicities (as not sure whether arrays are of same length).
    return squares_count == array2_count
```
