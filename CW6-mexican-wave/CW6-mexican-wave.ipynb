Introduction
The wave (known as the Mexican wave in the English-speaking world outside North America) is an example of metachronal rhythm achieved in a packed stadium when successive groups of spectators briefly stand, yell, and raise their arms. Immediately upon stretching to full height, the spectator returns to the usual seated position. The result is a wave of standing spectators that travels through the crowd, even though individual spectators never move away from their seats. In many large arenas the crowd is seated in a contiguous circuit all the way around the sport field, and so the wave is able to travel continuously around the arena; in discontiguous seating arrangements, the wave can instead reflect back and forth through the crowd. When the gap in seating is narrow, the wave can sometimes pass through it. Usually only one wave crest will be present at any given time in an arena, although simultaneous, counter-rotating waves have been produced.

In this simple Kata your task is to create a function that turns a string into a Mexican Wave. You will be passed a string and you must return an array of strings where an uppercase letter is a person standing up.

Ideas.

* Iterate through the string, taking progressively smaller slices from the left, and capitalise, then join together with slices.

```python
# Final submission.

def wave(people):
    lst = []
    if not people:
        return lst
    for i, char in enumerate(people):
        if char.isspace():
            continue
        else:
            lst.append(people[0: i] + people[i: len(people) + 1].capitalize())
    return lst
```

Code review.

Not so good - remember you want to move to a more Pythonic, declarative style of programming. Your code is quite imperative, and adds a lot of redundancy. The ideas you wrote down do suggest you are thinking less "atomistically", but notice that you are still writing in terms of how you want the program to exectue, rather than defining transformations via Python built-ins in terms of where you want to get to.

Deficiencies.

* The first guard clause you use is redundant - if `people` is an empty list, then the loop won't populate `lst` with anything anyway.
* Possibly redundant `enumerate` - given that you need indexing and slicing, and not the character itself, you can just use `range(len(people))`.
* No need to specify a slice end point for capitalise.
* Also, good you are thinking about Python built-ins, but use `.upper()` to capitalise just the individual letter, otherwise `hEllo`, would give `Hello`.

Solution.

* The proper syntax to specify this declaratively is to use list comprehensions, and then filter the list.
* If you find yourself writing `list.append()` - think list comprehension.

```python
def wave(people):
    return [
            people[0:i] + people[i].upper() + people[i+1:]
            for i in range(len(people))
            if people[i].isalpha()
    ]
```
