If we were to set up a Tic-Tac-Toe game, we would want to know whether the board's current state is solved, wouldn't we? Our goal is to create a function that will check that for us!

Assume that the board comes in the form of a 3x3 array, where the value is 0 if a spot is empty, 1 if it is an "X", or 2 if it is an "O", like so:

[[0, 0, 1],
 [0, 1, 2],
 [2, 1, 0]]
We want our function to return:

-1 if the board is not yet finished AND no one has won yet (there are empty spots),
1 if "X" won,
2 if "O" won,
0 if it's a cat's game (i.e. a draw).
You may assume that the board passed in is valid in the context of a game of Tic-Tac-Toe.

Ideas.

* A win state occurs when one of the columns/rows/diagonal consists of the same number.
    * In this instance Who has won is determined by the exact number.
* If no one has won, and there are empty spots, then it's unfinished
* If no one has won, and there aren't any empty spots, then it's a draw. 

```python
arr = [[1, 1, 1],
       [0, 2, 2],
       [0, 0, 0]]
```

```python
status = 0

# Form diagonals by treating game board positions as elements of a matrix.
# Right -> left diagonal has index (0,2), (1, 1), (2, 0)
diag_1 = [arr[j][j] for j in range(len(arr))]
diag_2 = [arr[id[0]][id[1]] for id in enumerate(range(2, -1, -1))]

for i, row in enumerate(arr):
    # Check rows, columns, diagonals for win state.
    col = [row[i] for row in arr]
    if (row.count(1) == 3) or (col.count(1) == 3) or (diag_1.count(1) == 3) or (diag_2.count(1) == 3):
        status = 1
        print(status)
        break
    elif (row.count(2) == 3) or (col.count(2) == 3) or (diag_1.count(2) == 3) or (diag_2.count(2) == 3):
        status = 2
        print(status)
        break

flat_arr = [arr[i][j] for i in range(len(arr)) for j in range(len(arr))]
if status != 1 and status !=2 and (0 in flat_arr):
    print(-1)
elif status != 1 and status !=2 and (0 not in flat_arr):
    print(0)
```

```python
# As a function, we can use can use return instead of break when any player wins.
# So no need to check if neither player has won when checking draw or incomplete state.

def is_solved(board):

    # Form diagonals by treating game board positions as elements of a matrix.
    # Right -> left diagonal has index (0,2), (1, 1), (2, 0)
    diag_1 = [board[j][j] for j in range(len(board))]
    diag_2 = [board[id[0]][id[1]] for id in enumerate(range(2, -1, -1))]

    for i, row in enumerate(board):
    # Check rows, columns, diagonals for win state.
        col = [row[i] for row in board]
        if (row.count(1) == 3) or (col.count(1) == 3) or (diag_1.count(1) == 3) or (diag_2.count(1) == 3):
            return 1
        elif (row.count(2) == 3) or (col.count(2) == 3) or (diag_1.count(2) == 3) or (diag_2.count(2) == 3):
            return 2
        
    flat_arr = [board[i][j] for i in range(len(board)) for j in range(len(board))]
    if 0 in flat_arr:
        return -1
    else:
        return 0
```

```python
board = [[1, 1, 1],
         [0, 2, 2],
        [ 0, 0, 0]]

is_solved(board)
```
