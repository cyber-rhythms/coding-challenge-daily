Write a function that takes a positive integer and returns the next smaller positive integer containing the same digits.

For example:

next_smaller(21) == 12
next_smaller(531) == 513
next_smaller(2071) == 2017
Return -1 (for Haskell: return Nothing, for Rust: return None), when there is no smaller number that contains the same digits. Also return -1 when the next smaller number with the same digits would require the leading digit to be zero.

next_smaller(9) == -1
next_smaller(135) == -1
next_smaller(1027) == -1  # 0721 is out since we don't write numbers with leading zeros
some tests will include very large numbers.
test data only employs positive integers.
The function you write for this challenge is the inverse of this kata: "Next bigger number with the same digits."

Ideas & notes.

* If we view a fixed number 1234 as a series of boxes, then given that we can only use these digits, the largest and smallest is given by placing largest digit in 1st box, etc.
* Examples suggest permuting the last two digits - are there any exceptions. No. Permuting pneultimate and last digit results in max diff of 81. Permuting... in max diff of 810 etc.
    * Think harder.
* Algebraically, we have $100a + 10b + c$.
* We want to permute the numbers to form the next smallest integer, so only $a$, $b$, $c$ can switch positions.
* $a > b > c$ is largest.
* Can solve this as a finnicky maths problem, but not so much for a CS problem
* Could just brute force generate and find the closest.
* One approximation we can use is to take the last digit, and switch with any leftmost number that is larger.
    * Now deal with edge cases.
    * What if the last digit is the largest?
* Can refine. 
    * Startng with index $n$ digit.
    * Look at the first $n-1$ digits.
    * Take the closest digit that is larger than index $n$ digit and swap.
        * Not quite a swap, rather, place index before that closest digit.

```python
# Take the closest digit that is larger than index $n$ digit and swap
# Do not want to modify in place.
n = 9702
digits = [int(digit) for digit in list(str(n))]
new_digits = digits.copy()
for i in range(len(digits) - 1, -1, -1):
    if digits[i] > digits[-1]:
        new_digits.insert(i, digits[-1])
        new_digits.pop()
        next_smaller = int("".join([str(digit) for digit in new_digits]))
        print(next_smaller)
        break
if new_digits[0] == 0 or digits == new_digits:
    print(-1)
```

Refinement.

* The above code fails for when $n = 414$.
* And that is because it takes the last index-3 digit, then scans the index-2 and index-1 digits, and finds that there are no larger digits, then returns $414, when actually $144$ is the answer.
* View $a_1a_2\dot a_n$ as an *ordered sequence* of elements.
* Instead, view the problem as finding the largest index $i$ such that $a_i > a_{i+1}$.
* Think about an example $n = 4122$.
* Your algorithm outputs $2412$, but actually the answer is $1422$, so don't start at the last digit. Instead, notice that the pivot point from which we run your algorithm actually starts at index-2.
* Got it - so this is about finding a suitable pivot from which to run your algorithm.
* Let's try another example, say $n = 87122$. Your code would output $82712$, when the answer is $82721$.
* So instead of starting at the last digit, we scan from right to left to find the first digit that is greater than the one to the right, i,e, $a_i$ such that 
$$a_i : a_{i-1} > a_{i}$$

```python
n = 87122
digits = [int(digit) for digit in list(str(n))]
new_digits = digits.copy()
for i in range(len(digits) - 1, -1, -1):
    if digits[i] > digits[-1]:
        new_digits.insert(i, digits[-1])
        new_digits.pop()
        next_smaller = int("".join([str(digit) for digit in new_digits]))
        print(next_smaller)
        break
if new_digits[0] == 0 or digits == new_digits:
    print(-1)
```
