For a given string s find the character c (or C) with longest consecutive repetition and return:

(c, l)
where l (or L) is the length of the repetition. If there are two or more characters with the same l return the first in order of appearance.

For empty string return:

('', 0)

Ideas.

* Any declarative ways to do this jump out a tyou?
* A count dictionary wouldn't work, as this would only count occurrence of a single letter, and destroy info about repetition length.
* Idea 1: Will need to do an $O(n)$ linear search of the entire string, so iterate through.
* For each character, we check compare current and next char; and if they are the same we add the current char and increment 1 to a char-repetition state list, otherwise clear the running char-repetition list.
    - But we would need another state variable to store the current char-repetition maximum right?
* Idea 2: Would a dictionary here help? Could run through an construct index: (char-repetition), then take a maximum?
* Either way would be $O(n)$ time, although dictionary would be two passes at $O(n)$ due to maximum, amd would incur additional $O(n)$ space.
* Looks like a slight twist on finding maximum of an array - where there is more state tracking.

```python
def longest_repetition(chars):
    if not chars:
        return ("", 0)

    max = [chars[0], 1]
    current = [chars[0], 1]

    for i in range(len(chars) - 1):

        # Update running character-repetition state list.
        if chars[i] == chars[i + 1]:
            current[1] += 1

            # Compare with maximum character-repetition state, if there is an update.
            if current[1] > max[1]:
                max[0], max[1] = chars[i], current[1]

        # Reset the running character-repetition state if next char is distinct.
        else:
            current[0], current[1] = chars[i + 1], 1

    return max[0], max[1]
```
