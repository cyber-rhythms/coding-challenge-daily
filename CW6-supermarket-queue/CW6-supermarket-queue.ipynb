There is a queue for the self-checkout tills at the supermarket. Your task is write a function to calculate the total time required for all the customers to check out!

input
customers: an array of positive integers representing the queue. Each integer represents a customer, and its value is the amount of time they require to check out.
n: a positive integer, the number of checkout tills.
output
The function should return an integer, the total time required.

Important
Please look at the examples and clarifications below, to ensure you understand the task correctly :)

Examples

```
queue_time([5,3,4], 1)
# should return 12
# because when n=1, the total time is just the sum of the times

queue_time([10,2,3,3], 2)
# should return 10
# because here n=2 and the 2nd, 3rd, and 4th people in the 
# queue finish before the 1st person has finished.

queue_time([2,3,10], 2)
# should return 12
```

Clarifications
There is only ONE queue serving many tills, and
The order of the queue NEVER changes, and
The front person in the queue (i.e. the first element in the array/list) proceeds to a till as soon as it becomes free.
N.B. You should assume that all the test input will be valid, as specified above.

Ideas.

* Not sure what data structure to choose here, do I need a list of list?
* Some instinct suggests modulo operations but I can't quite put my finger on why. 
* Idea 1: Think of each of the $n$ tills as boxes, which could be positions in a list.
* If you have $n$ customer durations and there is $1$ till, then total duration is a simple sum.
* At the beginning the first $n$ customers simply go to $n$ tills - we allocate them accordingly.
* For person $n+1$ in the queue, we now allocate them to the till with the lowest duration, as that will finish first - accumulating the duration in each till position.
    * What if there are multiple tills with the lowest duration? How do we tie break? Doensn't appear to matter.
* The time it takes is then the maximum across the accumulated till durations.
* Looks like it needs to be coded imperatively - I can't think of any declarative pattern.

````python
# Submission.

def queue_time(customers, n):
    if n == 1:
        return sum(customers)
    else:
        # Allocate first $n$ customers to $n$ tills.
        tills = customers[0:n]

        # Now run through remaining and allocate accordingly.
        for customer in customers[n:]:
            shortest_time = min(tills)
            min_index = next(i for i in range(len(tills)) if tills[i] == shortest_time)
            tills[min_index] += customer

        # Take a maximum of the till durations. 
        queue_time = max(tills)
        return queue_time
````

Code Review.

Not bad, but verbose. Given that you forgot how to retrieve the index of the minimum/maximum using built-ins, making use of a generator to do this is a good use of the tools you had available to you at the time.

Deficiencies.
* Starting with an empty list of zeros simplifies the slicing logic, and the edge ase handling - in a way you were held back computationally by the ideas you articulated (from a human perspective of standing in a queue)

Solution.
* See below - there are many ways to find the index of a max/min/sorted using `key=` - observe this one and write up.

````python
# Code review.

def queue_time(customers, n):
    tills = [0] * n
    for customer in customers:
        min_index = min(range(len(tills)), key=tills.__getitem__)
        tills[min_index] += customer
    return max(tills)
````
