In this kata you are required to, given a string, replace every letter with its position in the alphabet.

If anything in the text isn't a letter, ignore it and don't return it.

"a" = 1, "b" = 2, etc.

Example
Input = "The sunset sets at twelve o' clock."
Output = "20 8 5 19 21 14 19 5 20 19 5 20 19 1 20 20 23 5 12 22 5 15 3 12 15 3 11"

Ideas.

* Create a dictionary of letter-alphabet positions.
* Modify the string so that everything is lower.
* Use a list comprehension and then join all looked up alphabet positions, rather than joining the alphabet positions in a loop (strings are immutable, lists contain pointers)

```python
string = "The narwhal bacons at midnight.".lower()
alphabet = "abcdefghijklmnopqrstuvwxyz"
pos = {letter: str(i) for i, letter in enumerate(alphabet, start=1)}
" ".join([pos.get(char) for char in string if char in pos])
```
