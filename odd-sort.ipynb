Task
You will be given an array of numbers. You have to sort the odd numbers in ascending order while leaving the even numbers at their original positions.

Examples
[7, 1]  =>  [1, 7]
[5, 8, 6, 3, 4]  =>  [3, 8, 6, 5, 4]
[9, 8, 7, 6, 5, 4, 3, 2, 1, 0]  =>  [1, 8, 3, 6, 5, 4, 7, 2, 9, 0]

```python
arr = [9, 8, 7, 6, 5, 4, 3, 2, 1, 0] 
```

```python
# Store odd number positions in a dictionary.
index_num = {i: number for i, number in enumerate(arr) if number % 2 == 1}
index_num
```

```python
# Sort the values of the dictionary, and put back in.
sorted(index_num.values())
```

```python
# Sort the odd values in the dictionary, and use the index keys to put back into the array.
for index, value in zip(index_num.keys(), sorted(index_num.values())):
    print(index, value)
    arr[index] = value
```

```python
# A more Pythonic way is to only construct a list of odds, then iterate through them and rebuild the list. No need for a dictionary, just use iterators and lists.

def sort_array(source_array):

    # Create a list of odds.
    odds = sorted([x for x in source_array if x % 2 != 0])

    # Get iterator from list of odds.
    odds_iter = iter(odds)

    # Create a new list by running through original, then inserting a sorted odd if odd, and original number if even.
    return [next(odds_iter) if x % 2 != 0 else x for x in source_array]
```
