Implement the function unique_in_order which takes as argument a sequence and returns a list of items without any elements with the same value next to each other and preserving the original order of elements.

For example:

unique_in_order('AAAABBBCCDAABBB') == ['A', 'B', 'C', 'D', 'A', 'B']
unique_in_order('ABBCcAD')         == ['A', 'B', 'C', 'c', 'A', 'D']
unique_in_order([1, 2, 2, 3, 3])   == [1, 2, 3]
unique_in_order((1, 2, 2, 3, 3))   == [1, 2, 3]

https://www.codewars.com/kata/54e6533c92449cc251001667/train/python

Ideas.

* If make use of a set to get unique elements, then that will destroy the ordering of the list - so that's a no.
* Sliding window of length three could have more complex logic than needs to be.
* Maybe I'm making this more complex than it needs to be -- a list comprehension that compares to the previous character will be sufficient - use slices as they are nore robust for previous letter comparisons than array indexing.

```python
# Submission.

def unique_in_order(seq):
    return [seq[i:i+1][0] for i in range(0, len(seq)) if seq[i:i+1] != seq[i-1:i]]
```

```python
# Code review.

def unique_in_order(iterable):
    result = []
    prev = None
    for char in iterable[0:]:
        if char != prev:
            result.append(char)
            prev = char
    return result
```
