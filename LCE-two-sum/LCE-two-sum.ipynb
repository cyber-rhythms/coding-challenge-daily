```python
# Submission.
# Can only see an $O(n^2 / 2) = O(n^2)$ time, $O(1)$ space brute-force solution for this. - write it out first.

class Solution:
    def brute_twoSum(self, nums: list[int], target: int) -> list[int]:
        n = len(nums)

        # Iterate through each element in the array.
        for i in range(n):
            # Iterate through subsequent elements in the array, excluding same element pairs.
            for j in range(i+1, n):
                if nums[i] + nums[j] == target:
                    return [i, j]
        return []
```

**HINT 1 - Additive complements.**
* The proper solution is to get a solution with less than $O(n^2)$ time complexity.
* Think about additive complements and hash maps/dictionaries.
* Using an additive complement would still mean that we need to run an $O(n^2)$ time search - for each of $n$ array element additive complements, we search $O(n-1)$ elements and compare.

**HINT 2: - Hash-maps for backward looking O(1) lookup.**

* The key idea here is to trade time for memory - using a hash map.
* But what do we hash?
    * What are we searching for? We are searching for a complement.
    * What information do we need to retrieve once found? We require the index of said complement.
* The hash map will consist of `{array element: index}` pairs - because we need to retrieve indices, given array elements.
* We run one single-pass through $n$ elements of the array. For each element, we compute the complement, and if the complement is a key in the hash map, return current index, and also index of the looked up complement.
* Otherwise, add the current array element to the hash-map together with its index.
* In this way, we are iterating through $n$ elements, performing an $O(1)$ lookup, for a total of $O(n)$ time complexity, and $O(n)$ space complexity.
* The graduation from additive complement and $O(n^2)$ search, to this solution, is as follows.
    * Iterate through $n$ elements and compute an additive complement.
    * Checking whether the additive complement is in the hash-map is a more effiicent means of running a backward looking $O(n)$ linear search of all previous array elements.
    * Storing the array element and index number in the hashmap upgrades the array data structure with a memory, resulting in a time-complexity of saving from $O(n)$ linear search to $O(1)$ lookup.

```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        hash_map = dict()
        for index, num in enumerate(nums):
            complement = target - num

            # O(1) lookup of complement in hash-map rather than an O(n) linear search. 
            if complement in hash_map:
                index_2 = hash_map[complement]
                return [index, index_2]
            
            # Add current element-index pair to hash-map if complement not found.
            hash_map[num] = index
        return []
```
