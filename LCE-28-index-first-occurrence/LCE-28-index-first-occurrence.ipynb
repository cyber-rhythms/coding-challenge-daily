* Key part of LeetCode is that if you have what looks like a declarative Pytjon one-liner - you need to supply an imperative program - because there may be speedups/it may test a particular DSA pattern.
* Here, the declarative one-liner is `haystack.find(needle)`.
* Here it looks like we need to code a simple sliding window - write it out on paper, then adjust for indexing.
* Added the calculation to the coding journal as you had an off-by-one error.

```python
haystack = "butsadwxy"
needle = "sad"
```

```python
# Submission.

class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        n = len(haystack)
        k = len(needle)
        for i in range(n-k+1):
            current_window = haystack[i:i+k]
            if needle == current_window:
                return i
        return -1
```

```python
sol = Solution()
sol.strStr(haystack, needle)
```

* Slicing doesn't really count as imperative programming - it masks loops underneath.

```python
# My imperative version.

def strStr(haystack, needle):
    n = len(haystack)
    k = len(needle)

    for i in range(n-k+1):
        matched_count = 0
        for j in range(k):
            if haystack[i + j] == needle[j]:
                matched_count += 1
            else:
                break
        else:
            return i
    return -1
```
