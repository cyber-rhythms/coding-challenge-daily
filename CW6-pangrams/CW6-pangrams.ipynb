A pangram is a sentence that contains every single letter of the alphabet at least once. For example, the sentence "The quick brown fox jumps over the lazy dog" is a pangram, because it uses the letters A-Z at least once (case is irrelevant).

Given a string, detect whether or not it is a pangram. Return True if it is, False if not. Ignore numbers and punctuation.

https://www.codewars.com/kata/545cedaa9943f7fe7b000048/train/python

Ideas.

* Have a reference string, which you can turn into a set.
* Use a set to filter out unique alphabetic characters.
* Compare membership of reference set to string set, using a list of Booleans, then check for all true.

```python
# Final submission.

def is_pangram(string):
    alphabet = set("abcdefghijklmnopqrstuvwxyz")
    string = "ABCD45EFGH,IJK,LMNOPQR56STUVW3XYZ"
    chars = {char.lower() for char in string if char.isalpha()}
    membership = [1 if letter in chars else 0 for letter in alphabet]
    return len(membership) == 26 and all(membership)
```
