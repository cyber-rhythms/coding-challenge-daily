Ideas.

* Brute-force method $O(n)$ time, $O(n)$ space.
    * Transform integer into something we can iterate over (either a string or a list), of length $n$.
    * Simultaneously iterate going from index $0$ to $n-1$, and also index $n-1$ to $0$, and compare elements.
    * So we loop through entire array from $i = 0, 1, \dots, n-1$ and compare element $x_i$ with $x_{(n-1)-i}$.
    * Using lazy sequence/generator may mean space efficiency, instead of forming reversed lists?

```python
class Solution:
    def brute_isPalindrome(self, x: int) -> bool:
        digits = list(str(x))
        for digit, digit_2 in zip(digits, reversed(digits)):
            if digit != digit_2:
                return False
        return True
```

* Refining the idea.
* The above smells like a two-pointers technique - because we are simultaneously traversing iterable from left to right, and right to left.
    * Haven't studied two pointers in depth, only seen it in passing in binary search.
    * But the key idea is that after the mid-point of the array, we are duplicating work that doesn't need to be done.
    * This is the right way to go -- but how do we deal with the case where there is an odd/even number of digits?
        * This could be achieved with indefinite iteration, and odd/even issue can be circumvented with good coding of the while loop.

* Two pointers method $O(n)$ time, $O(n)$ space.
    * Set left and right pointers to be indices $0$ and $n-1$.
    * Then while `left` $\leq$ `right`, then run comparisons of the integers.
    * Terminate while loop if any comparison is False, so number isn't palindrome.
    * Otherwise if while loop successfully completes, the True. 

```python
# Final submission.

class Solution:
    def isPalindrome(self, x: int) -> bool:
        left = 0
        right = len(x) - 1
        while left <= right:
            if x[left] != x[right]:
                return False
            left += 1
            right -= 1
        return True
```
