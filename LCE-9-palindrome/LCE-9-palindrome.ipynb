Ideas.

* Brute-force method $O(n)$ time, $O(n)$ space.
    * Transform integer into something we can iterate over (either a string or a list), of length $n$.
    * Simultaneously iterate going from index $0$ to $n-1$, and also index $n-1$ to $0$, and compare elements.
    * So we loop through entire array from $i = 0, 1, \dots, n-1$ and compare element $x_i$ with $x_{(n-1)-i}$.
    * Using lazy sequence/generator may mean space efficiency, instead of forming reversed lists?

```python
class Solution:
    def brute_isPalindrome(self, x: int) -> bool:
        digits = list(str(x))
        for digit, digit_2 in zip(digits, reversed(digits)):
            if digit != digit_2:
                return False
        return True
```

* Refining the idea.
* The above smells like a two-pointers technique - because we are simultaneously traversing iterable from left to right, and right to left.
    * Haven't studied two pointers in depth, only seen it in passing in binary search.
    * But the key idea is that after the mid-point of the array, we are duplicating work that doesn't need to be done.
    * This is the right way to go -- but how do we deal with the case where there is an odd/even number of digits?
        * This could be achieved with indefinite iteration, and odd/even issue can be circumvented with good coding of the while loop.

* Two pointers method $O(n)$ time, $O(n)$ space.
    * Set left and right pointers to be indices $0$ and $n-1$.
    * Then while `left` $\leq$ `right`, then run comparisons of the integers.
    * Terminate while loop if any comparison is False, so number isn't palindrome.
    * Otherwise if while loop successfully completes, the True. 

```python
# Final submission.

class Solution:
    def isPalindrome(self, x: int) -> bool:
        left = 0
        right = len(x) - 1
        while left <= right:
            if x[left] != x[right]:
                return False
            left += 1
            right -= 1
        return True
```

* However, it's still not entirely clear how to do this without converting the integer into an iterable in memory like a list/string?
* Two-pointers does not improve on brute-force unless we find a way to make the integer `x` iterable, which requires turning into a string.
* Solution.
    * Observation: If $x = 1331$ is an integer of even length $n = 4$ then we only need to compare the first $n / 2$ digits, "the top half of $x$" with the second batch of $n/2$ digits, the "bottom half of $x$", where the order of the second batch of digits is reversed.
    * Tools observation: For an integer $x$, then `x % 10` returns the final $n$th digit of the integer, and `x // 10` returns the first $n-1$ digits of the integer.
    * This grants us a limited tool to access an integer *almost as if it were indexed*.
    * Beginning with integer `x` on the "left" side of a "balancing scale", and `bottom_half_x` initialised at 0 - imagine an unbalanced scale.
    * We now incrementally remove digits from `x` and transfer to the "right hand side of balancing scale", building the `bottom_half_x` - using the modulo tools.
        * Palindrome: $(1331, 0) \rightarrow (133, 1) \rightarrow (13, 13)$.
        * Non-palindrome $(1234, 0) \rightarrow (123, 4) \rightarrow (12, 43)$
    * Odd number of integers can be wrapped into the logic of indefinite iteration.
        * Palindrome: (13531, 0) -> (1353, 1) -> (135, 13) -> (13, 135) 
        * Non-palindrome (12345, 0) -> (1234, 5) -> (123, 54) -> (12, 543).

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
    
        # Filter out negative integers and multiples of 10.
        # Allow 0 through.
        if x < 0 or (x % 10 == 0 and x != 0):
            return False
        
        bottom_half_x = 0 
        while x > bottom_half_x:
            bottom_half_x = (bottom_half_x * 10) + (x % 10)
            x //= 10
        
        # Integers with an odd number 2n + 1 digits will end with x having n digits.
        # And bottom_half_x having n + 1 digits - so remove the last digit to run palindrome check.
        return (x == bottom_half_x or x == bottom_half_x // 10)
```
