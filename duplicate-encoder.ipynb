Ignore capitalization when determining if a character is a duplicate.

Examples
"din"      =>  "((("
"recede"   =>  "()()()"
"Success"  =>  ")())())"
"(( @"     =>  "))(("

Code review.

1. Best to use Collections.count to create a count dictionary.
2. Dictionary comprehensions cannot refer to themselves from within.
3. Need to formalise your intuition about global counts into algorithm analysis - if you'd used `count()` here it would have been $O(n^2)$.
4. On `.join()` strings in a `for` loop vs using a list comprehension and then `.join()`
    * Strings are immutable - better to create a whole list of strings with a mutable list comprehension and join, than incrementally join string components within a loop.
    * Memory-wise, the first option constructs 1000 distinct immutable strings in memory, whereas the second constructs 2 distinct "(" and ")" object is in memories with 1000 pointers.
    * Time-wise, the first option means that there are $1 + \dots + n = n(n+1)/2 \approx n^2$, so we hae write 1 char, write 2 chars, write 3 chars and so on, with $O(n^2)$ time complexirt whereas the second collects pointers in $O(n)$ time. 

```python
from collections import Counter

def duplicate_encode(word):
    word = word.lower()
    counts = Counter(word)
    return "".join(["(" if counts.get(char) == 1 else ")" for char in word])
```
