```python
n = 5
arr = map(int, ["2", "3", "6", "6", "5", "4"])

arr = list(arr)
ismax = lambda x: True if x == max(arr) else False
    
for score in sorted(arr, reverse=True):
    if not ismax(score):
        print(score)
        break
```

Code Review.

Deficiencies.

* Your thinking here was: Iterate through the sorted list, and the first number that isn't a maximum is the runner-up score.
* The problem here looks deceptively simple - but you've fallen into many traps in your first solution.
* Issue 1: `sorted()` is an $O(n)$ operation, and then computing `max()` is also $O(n)$, for a total of $O(n^2)$ time complexity.
* Issue 2: Storing a list is also $O(n)$ space complexity.

Solution.
* Remember - promote data sparsity - we don't care about any elements of the same magnitude, only unique elements - use a set.
* Instead think like this: Take the unique elements, sort them, retrieve the index of the next smallest.

```python
# Professional SWE version.

if __name__ == '__main__':
    n = int(input())
    arr = map(int, input().split())

    arr = sorted(list(set(arr)), reverse=True)
    runup_score = arr[1]
    print(runup_score)
```
