Count the number of Duplicates
Write a function that will return the count of distinct case-insensitive alphabetic characters and numeric digits that occur more than once in the input string. The input string can be assumed to contain only alphabets (both uppercase and lowercase) and numeric digits.

Example
"abcde" -> 0 # no characters repeats more than once
"aabbcde" -> 2 # 'a' and 'b'
"aabBcde" -> 2 # 'a' occurs twice and 'b' twice (`b` and `B`)
"indivisibility" -> 1 # 'i' occurs six times
"Indivisibilities" -> 2 # 'i' occurs seven times and 's' occurs twice
"aA11" -> 2 # 'a' and '1'
"ABBA" -> 2 # 'A' and 'B' each occur twice

Ideas.

* What data structure? Form a character-count dictionary, either by a dictionary comprehension or using collections counter, then iterate through and extract distinct characters.
* Failure mode 1: Empty string.
* Easy case: No repeats.

```python
from collections import Counter
text = "Indivisibilities"

def duplicate_count(text):
    text = text.lower()
    if not len(text):
        return 0
    else:
        counts = Counter(text)
        distinct_chars = [char for char in counts if counts.get(char) > 1]
        return len(distinct_chars)
```

Code review.

* Use of collections `Counter` is the proper way to construct the dictionary here.
* Can make the list comprehension here more memory efficient, as we bin the list, and only take its length. Use a generator instead:

```python
def duplicate_count(text):
    counts = Counter(text.lower())
    return sum(1 for count in counts.values() if count > 1)
```
