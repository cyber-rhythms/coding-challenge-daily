* A simple, ad-hoc way to do this, without being clever:
    * Use `collections Counter` to create a dictionary of counts of each letter.
    * Sort the dictionary by value - if it isn't sorted in descending frequency order as a collections Counter already?
    * Can use `counter.most_common(n)` for the `n` most common `key-values` in the counter dictionary.

- HINT: Issue if you use `counter.most_common(n)` is that if you have ties on letter counts involving more than three letters, using the `counter.most_common(3)` method will mean that you will select an arbitrary slice in the orders they were added, and not alphabetically.
    * The difficulty here is that you have more than one level of sorting going on - you have to coordinate a frequency sort AND an alphabetic sort
    * If you code this imperatively - try alphabetically *first*, then sort *by frequency*, then take a slice - try this, then can pick up with a declarative solution (which builds on this).

```python
# Submission.
# When you cover sorting algorithms in DSA, and multi-pass sorting, make sure you have a look at sorting algorithm "stability".

from collections import Counter

if __name__ == '__main__':
    s = input()
    counts = Counter(s)
    
    # Sort alphabetically first, then by frequency.
    alphabetic_order = sorted(counts.items())
    freq_order = sorted(alphabetic_order, key=lambda x: x[1], reverse=True)
    
    for letter, count in freq_order[:3]:
        print(f"{letter} {count}")
```
