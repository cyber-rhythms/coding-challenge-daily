Given a string of words, you need to find the highest scoring word.

Each letter of a word scores points according to its position in the alphabet: a = 1, b = 2, c = 3 etc.

For example, the score of abad is 8 (1 + 2 + 1 + 4).

You need to return the highest scoring word as a string.

If two words score the same, return the word that appears earliest in the original string.

All letters will be lowercase and all inputs will be valid.

```python
x = 'what volcano time are yyyg we climbing ul the'
```

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
letter_score = {letter: i + 1  for i, letter in enumerate(alphabet)}
letter_score
```

```python
word_score = 0
for letter in "what":
    word_score += letter_score[letter]
```

```python
def word_score(word):
    word_score = 0
    for letter in word:
        word_score += letter_score[letter]
    return word_score
```

```python
[word_score(word) for word in x.split()]
```

```python
scores = {word: (word_score(word), i) for i, word in enumerate(x.split())}
scores
```

```python
for key, value in scores.items():
    print(key, value)
```
