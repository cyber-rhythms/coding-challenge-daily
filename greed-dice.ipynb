Greed is a dice game played with five six-sided dice. Your mission, should you choose to accept it, is to score a throw according to these rules. You will always be given an array with five six-sided dice values.

 Three 1's => 1000 points
 Three 6's =>  600 points
 Three 5's =>  500 points
 Three 4's =>  400 points
 Three 3's =>  300 points
 Three 2's =>  200 points
 One   1   =>  100 points
 One   5   =>   50 point
Each of 5 dice can only be counted once in each roll. For example, a given "5" can only count as part of a triplet (contributing to the 500 points) or as a single 50 points, but not both in the same roll.

Example scoring

 Throw       Score
 ---------   ------------------
 5 1 3 4 1   250:  50 (for the 5) + 2 * 100 (for the 1s)
 1 1 1 3 1   1100: 1000 (for three 1s) + 100 (for the other 1)
 2 4 4 5 4   450:  400 (for three 4s) + 50 (for the 5)
Note: your solution must not modify the input list.

```python
from collections import Counter

def score(dice):
    points = 0
    # One pass to count everything: O(n)
    counts = Counter(dice)
    
    # Rule definitions
    triplet_values = {1: 1000, 6: 600, 5: 500, 4: 400, 3: 300, 2: 200}
    
    for face, count in counts.items():
        # Use floor division to find how many triplets exist
        num_triplets = count // 3
        if num_triplets > 0:
            points += num_triplets * triplet_values.get(face, 0)
        
        # Use modulo to find remaining singles
        remaining = count % 3
        if face == 1:
            points += remaining * 100
        elif face == 5:
            points += remaining * 50
            
    return points
```
