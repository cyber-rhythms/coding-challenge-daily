Enough is enough!
Alice and Bob were on a holiday. Both of them took many pictures of the places they've been, and now they want to show Charlie their entire collection. However, Charlie doesn't like these sessions, since the motif usually repeats. He isn't fond of seeing the Eiffel tower 40 times.
He tells them that he will only sit for the session if they show the same motif at most N times. Luckily, Alice and Bob are able to encode the motif as a number. Can you help them to remove numbers such that their list contains each number only up to N times, without changing the order?

Task
Given a list and a number, create a new list that contains each number of list at most N times, without reordering.
For example if the input number is 2, and the input list is [1,2,3,1,2,1,2,3], you take [1,2,3,1,2], drop the next [1,2] since this would lead to 1 and 2 being in the result 3 times, and then take 3, which leads to [1,2,3,1,2,3].
With list [20,37,20,21] and number 1, the result would be [20,37,21].

https://www.codewars.com/kata/554ca54ffa7d91b236000023/train/python

Ideas.

* Construct a count dictionary that progressively gets populated as the original list is traversed, then construct new list based on these running counts.
* Note that $N >= 1$ otherwise your list will contain no elements.
* Is it possible if I implement this to keep it Pythonic using just a list comprehension? Not sure, you have to keep a dictionary of running counts, otherwise repeatedly call `count` on the entire list which is inefficient.

```python
# Final submission.

def delete_nth(lst, N):
    counts = dict()
    new_list = []
    for num in lst:
        counts[num] = counts.get(num, 0) + 1
        if counts.get(num) <= N:
            new_list.append(num)
        else:
            continue
    return new_list
```

Code review.

Deficiencies.
* Right idea on using a count dictionary, but some syntactic redunancies you need to be aware of.
* Logic in the `if` - somewhat redundant - you could have referenced `counts.get(num, 0) + 1` to prevent a further dicit    oary lookup.
`else: continue` is redundant.

Solution.
* You can construct an empty count dictionary from collections using `defaultdict(int)` to automatically initialise missing keys with `0`, so you don't need `counts.get(num, 0)`
* Logic is cleaner if you do a check first.
* CodeWars most upvoted solution is a one-liener, but repeatedly calls `count()` on the whole list - not what I want to be doing.

```python
from collections import defaultdict

def delete_nth(lst, N):
    result = []
    counts = defaultdict(int)
    for x in lst:
        if counts[x] < N:
            result.append(x)
            counts[x] += 1

    return result
```
