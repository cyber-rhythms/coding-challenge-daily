```python
# Final submission.

if __name__ == '__main__':
    
    records = []
    scores = []
    
    for _ in range(int(input())):
        name = input()
        score = float(input())
        records.append([name, score])
        scores.append(score)
        
    unique_scores = sorted(list(set(scores)))
    penul_score = unique_scores[1]
    
    lazy_students = sorted(record[0] for record in records if record[1] == penul_score)
    
    for student in lazy_students:
        print(student)
```
