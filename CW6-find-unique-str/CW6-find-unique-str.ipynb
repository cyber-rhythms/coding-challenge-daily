There is an array of strings. All strings contains similar letters except one. Try to find it!

find_uniq([ 'Aa', 'aaa', 'aaaaa', 'BbBb', 'Aaaa', 'AaAaAa', 'a' ]) # => 'BbBb'
find_uniq([ 'abc', 'acb', 'bac', 'foo', 'bca', 'cab', 'cba' ]) # => 'foo'
Strings may contain spaces. Spaces are not significant, only non-spaces symbols matters. E.g. string that contains only spaces is like empty string.

It’s guaranteed that array contains more than 2 strings.

Ideas.

* As no distinction between capitals and lower case, lower case every string.
* Only composition of a string is important to assess if it is the common one, or the unique string - so turning every string into a set/signature will result in a list of set/signatures, one will be the odd one out.
* Unique string can be in the first or second position as an edge case, otherwise run a linear pass through the array to find it.

```python
# Submission.

def find_uniq(arr):
    
    signature_0, signature_1 = set(arr[0].lower()), set(arr[1].lower())
    
    if signature_0 != signature_1:
        signature_2 = set(arr[2].lower())
        return arr[0] if signature_0 != signature_2 else arr[1]

    common = signature_0
    for string in arr[2:]:
        signature = set(string.lower())
        if signature != common:
            return string
```
