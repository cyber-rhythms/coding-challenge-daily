""  -->  ""

```python
sentence = "is2 Thi1s T4est 3a"
if not sentence:
    print("")
else:
    words = sentence.split()
    sorted = [0] * len(words)
    nums = [str(i) for i in range(1, 10)]

    for i, word in enumerate(words):
        for char in word:
            if char in nums:
                index = int(char) - 1
        sorted[index] = word
    print(sorted)
    
    sorted_sent = " ".join(sorted)
    print(sorted_sent)
```
