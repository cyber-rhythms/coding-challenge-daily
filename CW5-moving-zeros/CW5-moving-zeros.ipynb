* For now, I would approach relatively naively.
* Idea 1: Expensive memory way would be to silo non zeros into one list and zeros into another and then join - $O(n)$ space and $O(n)$ time. I know this to be safe.
* Idea 2: Or could just use append and deletions for in-place modification at $O(1)$ space and $O(n)$ time -- but then surely you're not supposed to modify a list in place if you are also looping over it?
    * Think this is where DSA comes in.

```python
lst =  [9, 0, 0, 9, 1, 2, 0, 1, 0, 1, 0, 3, 0, 1, 9, 0, 0, 0, 0, 9]

nums = []
zeros = []

for num in lst:
    if num != 0:
        nums.append(num)
    else:
        zeros.append(0)

print(nums + zeros)
```
