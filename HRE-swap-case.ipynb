You are given a string and your task is to swap cases. In other words, convert all lowercase letters to uppercase letters and vice versa.

For Example:

Www.HackerRank.com → wWW.hACKERrANK.COM
Pythonist 2 → pYTHONIST 2  
Function Description

Complete the swap_case function in the editor below.

swap_case has the following parameters:

string s: the string to modify
Returns

string: the modified string
Input Format

A single line containing a string .

Constraints


Sample Input 0

HackerRank.com presents "Pythonist 2".
Sample Output 0

hACKERrANK.COM PRESENTS "pYTHONIST 2".

https://www.hackerrank.com/challenges/swap-case/problem?isFullScreen=true

```python
string = 'HackerRank.com presents "Pythonist 2".'

def swap_case(string):
    string_list = []

    for substring in string.split():
        substring_list = list(substring)

    for j, char in enumerate(substring_list):
        if char.isupper():
            substring_list[j] = char.lower()
        elif char.islower():
            substring_list[j] = char.upper()

    mod_substring = "".join(substring_list)
    string_list.append(mod_substring)

    return " ".join(string_list)
```

Code Review.

Deficiencies.

* The above is way too verbose.
* Note that firstly, we can iterate over the string itself - white spaces are characters themselves.
* For now, keep writing your ideas so you can quickly debug where your thinking went wrong.
* You mistakenly thought that you couldn't iterate over whitespace, or run character checks on whitespace - whitespace is a "character" in a string.
* For this reason, you tried to split the word into substrings to "preserve whitespace".
* Then you correctly reasoned that because strings are immutable, you needed a list to operate on substrings, and then this creates an extra overhead of joining the newly created strings.

Solution.
* The first and second loop can be merged into one clean piece of code using a generator, or you can use the built-in string method, `str.swapcase()`.

```python
string = 'HackerRank.com presents "Pythonist 2".'

def swap_case(string):
    return "".join(char.upper() if char.islower() else char.lower() for char in string)
```

```python
swap_case('HackerRank.com presents "Pythonist 2".')
```
