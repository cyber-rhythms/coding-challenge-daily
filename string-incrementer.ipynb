Your job is to write a function which increments a string, to create a new string.

If the string already ends with a number, the number should be incremented by 1.
If the string does not end with a number. the number 1 should be appended to the new string.
Examples:

foo -> foo1

foobar23 -> foobar24

foo0042 -> foo0043

foo9 -> foo10

foo099 -> foo100

Attention: If the number has leading zeros the amount of digits should be considered.

Ideas.

* A string either has a number at the end or it doesn't.
* The "nice" case is where it doesn't - we simply add a 1.
* Bulk of the logic will be in the case where it does have a number.
    * In order to properly increment this number by 1, we have to remove all leading 0s, if there are any, and increment by 1.
* Failure mode 1: When the string is empty, increment by 1.
* Failure mode 2: Numbers may not necessarily be at the end, even with leading zeros, to begin with, and there may be other numbers inside the string.

Coding logic.

* Iterate through the string.
* If it is empty, return a 1 (guard)
* If last letter is a string, add a 1 to the end.
* Otherwise, we find a number- > letter state transition, beginning at the end.
* We then isolate the string, by stripping all the leading 0s, and then incrementing.

```python
def increment_string(string):

    # Empty string exit.
    if not len(string):
        return "1"
    
    # No number at end exit.
    if string[-1].isalpha():
        return (f"{string}{1}")
    
    # Traverse string in reverse, find first number -> letter transition.
    for i in range(len(string) - 1, 0, -1):
        is_number = string[i].isnumeric()
        is_letter = string[i - 1].isalpha()

        if is_number and is_letter:
            incr_num = int(string[i:].strip("0")) + 1
            return f"{string[:i]}{incr_num}"
```

* If there are leading zeros, then we need to check whether incrementing the number by 1 causes a "base change", e,g. 9 -> 10, 99- > 100.
    * If there is a base change, then we have to remove one zero.
    * Otherwise, we retain the same number of 0s.
* But retaining and removing the zeros requires us knowing how many zeros there are at the beginning.

```python
string = "foobar000100"

for i in range(len(string) - 1, 0, -1):
    is_number = string[i].isnumeric()
    is_letter = string[i - 1].isalpha()

    if is_number and is_letter:

        # Check if there are leading zeros, this finds the first
        if string[i] == "0":

            # Compute the increment, check if there is a base change.
            num = int(string[i:].strip("0"))
            incr_num = num + 1
            base_change = len(str(incr_num)) - len(str(num))

            # Modify the string accordingly.
            stop = len(string) - len(str(incr_num)) - base_change

            print(f"{string[:stop]}{incr_num}")

        else:
            incr_num = int(string[i:].strip("0")) + 1
            print(f"{string[:i]}{incr_num}")
            break
```
