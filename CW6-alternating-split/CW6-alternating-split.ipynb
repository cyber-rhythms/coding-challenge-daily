Implement a pseudo-encryption algorithm which given a string S and an integer N concatenates all the odd-indexed characters of S with all the even-indexed characters of S, this process should be repeated N times.

Examples:

encrypt("012345", 1)  =>  "135024"
encrypt("012345", 2)  =>  "135024"  ->  "304152"
encrypt("012345", 3)  =>  "135024"  ->  "304152"  ->  "012345"

encrypt("01234", 1)  =>  "13024"
encrypt("01234", 2)  =>  "13024"  ->  "32104"
encrypt("01234", 3)  =>  "13024"  ->  "32104"  ->  "20314"
Together with the encryption function, you should also implement a decryption function which reverses the process.

If the string S is an empty value or the integer N is not positive, return the first argument without changes.

Ideas.

* For encryption, we need a declarative transformation of the data, rather than trying to loop through.
* Could use two filtered list comprehensions and join - but would that be faster than just looping through once and creating a list imperatively? Asymptotically it doesn't matter, but then time wise it does doesn't it?
* For decryption, we need to know the number of odd or even indices, and so we need the length. With $0$-indexing, then if $n$ is even, there are $n / 2$ odd and even indices, whereas if $n$ is odd, then there are $(n+1)/2$ even indices and $(n-1) / 2$ odd indices - can find a "nicer" analytic solution later - turns out you don't even need this!
* Just use a list comprehension with filter to generate all the odd and even indices up to $n$!
* And can we use recursion here?

```python
# Submission.

def encrypt(text, n):
    if not text or n < 1:
        return text
    
    ciph_text = text[:len(text)]
    for _ in range(n):
        odds = [char for i, char in enumerate(ciph_text) if i % 2 == 1]
        evens = [char for i, char in enumerate(ciph_text) if i % 2 == 0]
        ciph_text = "".join(odds + evens)

    return ciph_text


def decrypt(ciph_text, n):

    if not ciph_text or n < 1:
        return ciph_text

    odd_indices = [i for i in range(len(ciph_text)) if i % 2 == 1]
    even_indices = [i for i in range(len(ciph_text)) if i % 2 == 0]
    indices = odd_indices + even_indices

    for _ in range(n):
        text = [0] * len(ciph_text)
        for index, char in zip(indices, ciph_text):
            text[index] = char
        ciph_text = text
    
    return "".join(ciph_text)
```

Code snippet below - always work with copies rather than modifying in place with functions, unless you have good reason, remember never to modify an iterable in place whilst looping over it! Make a copy instead.

```python
# Decryption

ciph_text = "01234"
k = len(ciph_text)

odd_indices = [i for i in range(k) if i % 2 == 1]
even_indices = [i for i in range(k) if i % 2 == 0]
indices = odd_indices + even_indices

text = list(ciph_text)

for index, char in zip(indices, text):
    text[index] = char
text
```
