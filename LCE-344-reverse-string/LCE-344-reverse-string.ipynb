Ideas.

* Key constraint here is to do this in-place with O(1) space, rather than O(n) space.
* The declarative version of this is in Python is simply `s.reverse()`, but guessing question is not asking for a declarative solution.
* One of the only safe ways to mutate an iterable in place whilst looping over it is to use a two-pointers technique.

```python
s = ["h","e","l","l","o"]

left = 0
right = len(s) - 1

while left <= right:
    s[left], s[right] = s[right], s[left]
    left += 1
    right -= 1
```

```python
class Solution:
    def reverseString(self, s: list[str]) -> None:
        # Set left and right pointers at beginning and end of list.
        left = 0
        right = len(s) - 1

        while left <= right:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1
```
